```python
import json
import time
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import random_split
from torch.utils.data import DataLoader, TensorDataset, Subset

from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

device = torch.device('cuda:3' if torch.cuda.is_available() else 'cpu')
```

```python
with open("/home/qfu1/ML_crystring/ML_crystring/CleanData/token_to_number_combined.json", "r") as f:
    token_to_number_combined = json.load(f)
with open("/home/qfu1/ML_crystring/ML_crystring/CleanData/number_to_token_combined.json", "r") as f:
    number_to_token_combined = json.load(f)
```

```python
df = pd.read_json("/home/qfu1/ML_crystring/ML_crystring/CleanData/CleanData(Tokenized).json", orient="records")
df = df[df["band_gap"]-1e-2 > 0]
df.head()
```

```python
max_sequence_len = max([len(seq) for seq in df['tokenized']])
X = torch.tensor(pad_sequences(df['tokenized'], maxlen=max_sequence_len, padding="pre"), dtype=torch.long)
y = torch.tensor(np.array(df["band_gap"]).reshape(-1,1), dtype=torch.float)

print(X.shape, y.shape)
```

```python
vocab_size = len(token_to_number_combined)+1
emb_dim = 16
hidden_size = 32
batch_size=512
```

```python
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.2)


train_dataset = TensorDataset(X_train, y_train)
val_dataset = TensorDataset(X_val, y_val)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=True)
```

```python
class LSTMModel_R(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_size, num_layers=3):
        super(LSTMModel_R, self).__init__()
        self.embedding_layer = nn.Embedding(num_embeddings=vocab_size, embedding_dim=emb_dim)
        self.lstm_layer = nn.LSTM(input_size=emb_dim, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(hidden_size, 512),  
            nn.SiLU(),
            nn.Linear(512, 128),
            nn.SiLU(),
            nn.Linear(128, 1),
            nn.ReLU()
        )

    def forward(self, x):
        # torch.Size([batch_size, max_sequence_len])
        embedding = self.embedding_layer(x)
        rnn_output, (hidden_state, cell_state) = self.lstm_layer(embedding)
        last_output = rnn_output[:, -1, :]
        # torch.Size([batch_size, hidden_size])
        final_output = self.fc(last_output)
        # torch.Size([batch_size, 1])
        return final_output
```

```python
def trian(train_loader, val_loader, model, criterion, optimizer, epochs, mute=False):
    TrainLoss = []
    ValLoss = []
    model.to(device)
    start_time = time.time()
    
    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss = 0
        for batch_X, batch_y in train_loader:
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)
            
            optimizer.zero_grad()                
            predictions = model(batch_X)
            loss = criterion(predictions, batch_y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
    
        # Validation phase
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for X_val, y_val in val_loader:
                X_val = X_val.to(device)
                y_val = y_val.to(device)
                predictions = model(batch_X)
                loss = criterion(predictions, batch_y)
                val_loss += loss.item()

        train_loss = train_loss/len(train_loader)
        val_loss = val_loss/len(val_loader)
        
        TrainLoss.append(train_loss)
        ValLoss.append(val_loss)
        
        if not mute:
            if (epoch + 1) % 10 == 0:
                end_time = time.time()
                training_time = end_time - start_time
                print(f'Epoch [{epoch+1}/{epochs}] - Train Loss: {train_loss:.4f}, Val Loss: {loss:.4f}, Time: {training_time:.2f}')
            
    return model, TrainLoss, ValLoss
```

```python
lr=2e-3
wd=8e-4
epochs = 200
num_layers = 3
```

```python
model = LSTMModel_R(vocab_size=vocab_size, emb_dim=emb_dim, hidden_size=hidden_size, num_layers=num_layers)
criterion = nn.MSELoss(reduction='mean')
optimizer = torch.optim.Adam(model.parameters(), lr=lr,weight_decay=wd)
model, TrainLoss, ValLoss = trian(train_loader, val_loader, model, criterion, optimizer, epochs)
```

```python
torch.save(model, f'/home/qfu1/TrainedModels/LstmToBandGap_Regression_lr{lr}_wd{wd}_ep{epochs}_nl{num_layers}_bc{batch_size}.pth')
model = torch.load(f'/home/qfu1/TrainedModels/LstmToBandGap_Regression_lr{lr}_wd{wd}_ep{epochs}_nl{num_layers}_bc{batch_size}.pth', weights_only=False)
```

```python
plt.plot(TrainLoss, label='Train Loss')
plt.plot(ValLoss, label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs")
plt.show()
```

```python
model.eval()
# model = model.to(device)
with torch.no_grad():
    X_test = X_test.to(device)
    all_predictions = model(X_test).cpu()

model.eval()
with torch.no_grad():
    X_train = X_train.to(device)
    all_predictions_train = model(X_train).cpu()
```

```python
mse = mean_squared_error(y_test, all_predictions)
mae = mean_absolute_error(y_test, all_predictions)
r2 = r2_score(y_test, all_predictions)

print("Test Score:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R² Score: {r2:.4f}")

plt.figure(figsize=(8, 6))
plt.scatter(y_train, all_predictions_train, alpha=0.5, edgecolors='k', label='train data')
plt.scatter(y_test, all_predictions, alpha=0.5, edgecolors='k', label='test data')
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], color='red', linestyle='--', linewidth=2)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.grid()
plt.legend()
plt.show()
```

### hyperparameter tuning

```python
import itertools
from tqdm import tqdm
```

```python
param_grid = {
    "emb_dim": [16],
    "hidden_size": [32],
    "lr": [1e-2, 8e-3, 5e-3, 1e-3, 8e-4],
    "wd": [1e-3, 5e-3, 1e-4],
    "num_layers": [3,5]
}
param_combinations = list(itertools.product(*param_grid.values()))
```

```python
grid_search_results = {
    "para": [],
    "train_loss": [],
    "val_loss": [],
    "mse": [],
    "mae": [],
    "r2": []
}

for params in tqdm(param_combinations, desc="Grid Search Progress", unit="set"):
    emb_dim, hidden_size, lr, wd, num_layers = params
    param_key = f"emb={emb_dim}_hidden={hidden_size}_batch={batch_size}_lr={lr}_wd={wd}_epochs={epochs}_layers={num_layers}"
    
    model = LSTMModel(
        vocab_size=vocab_size, emb_dim=emb_dim, hidden_size=hidden_size, num_layers=num_layers
    )
    
    criterion = nn.MSELoss(reduction='mean')
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=wd)
    
    model, TrainLoss, ValLoss = trian(train_loader, val_loader, model, criterion, optimizer, epochs, mute=True)
    
    model.eval()
    with torch.no_grad():
        X_test = X_test.to(device)
        y_pred = model(X_test).cpu()
    
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    grid_search_results["para"].append(param_key)
    grid_search_results["train_loss"].append(min(TrainLoss))
    grid_search_results["val_loss"].append(min(ValLoss))
    grid_search_results["mse"].append(mse)
    grid_search_results["mae"].append(mae)
    grid_search_results["r2"].append(r2)
```
