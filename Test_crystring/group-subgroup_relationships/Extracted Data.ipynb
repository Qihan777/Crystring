```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
for i in range(1, 231):  # Loop from 1 to 230
    # Construct the file names
    html_file = f'Group_subgroup/{i}.html'
    csv_file = f'Group_subgroup/extracted_data_{i}.csv'

    # Load and parse each HTML file
    with open(html_file, 'r') as file:
        content = file.read()
    soup = BeautifulSoup(content, 'html.parser')

    # Find the specific table - assuming the required table is the second one
    table = soup.find_all('table')[1]

    # Extract the headers
    headers = [header.text.strip() for header in table.find_all('th')]

    # Extract the rows
    rows = []
    for row in table.find_all('tr')[1:]:  # skipping the header row
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        rows.append(cols)

    # Create a DataFrame and save it as a CSV file
    df = pd.DataFrame(rows, columns=headers)
    df.to_csv(csv_file, index=False)
```

### Wyckoff Positions

```python
for i in range(1, 231):
    url = "https://www.cryst.ehu.es/cgi-bin/cryst/programs//nph-wp-list?gnum="+str(i).zfill(3)+"&what=wpos"
    html_file_path = "Wyckoff Positions/"+str(i)+".html"
    with urllib.request.urlopen(url) as response:
        content = response.read()
        
    # Saving the HTML content to a file
    with open(html_file_path, 'wb') as file:
        file.write(content)
```

```python
def extract_table_data(table):
    rows = table.find_all('tr')
    data = []
    for row in rows:
        cols = row.find_all(['td', 'th'])
        cols = [ele.text.strip() for ele in cols]
        data.append(cols)
    return data
```

```python
def filter_table(table):
    filtered_table = []
    for row in table:
        if row[0].isdigit():
            if "\n" in row[3]:
                row[3] = row[3].replace("\n", "")
            filtered_table.append(row[:4])
    return filtered_table
```

```python
for i in range(1, 231):
    html_file = f'Wyckoff_Positions/{i}.html'
    csv_file = f'Wyckoff_Positions/extracted_data_{i}.csv'
    
    with open(html_file, 'r') as file:
        content = file.read()

    soup = BeautifulSoup(content, 'html.parser')
    table = extract_table_data(soup)[1:-6]

    columns = table[0]
    table = filter_table(table)
    df = pd.DataFrame(table,columns = columns)
    df.to_csv(csv_file, index=False)
```
