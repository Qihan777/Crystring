```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import spglib
from pymatgen.io.ase import AseAtomsAdaptor
from monty.serialization import loadfn
```

```python
entries = loadfn("../data/MP_data-50.json.gz")
```

```python
for id in entries.keys():
    MP_structure = AseAtomsAdaptor.get_atoms(entries[id]["structure"])
    lattice = MP_structure.get_cell()
    positions = MP_structure.get_scaled_positions()
    numbers = MP_structure.get_atomic_numbers()
    cell = (lattice, positions, numbers)
    
    spacegroup = spglib.get_spacegroup(cell, symprec=1e-05, symbol_type=0)
    name, number = spacegroup.split(' ')
    symm = spglib.get_symmetry_dataset(cell)

    print(id,name,number[1:-1])
    print(symm["hall_number"])
```

```python
data = pd.read_csv("spglib_wyckoff.csv")
Info = data.loc[data['Hall number'] == symm["hall_number"]].reset_index(drop=True)["Info"][0]
print("Info:",Info[1:])
newdata = data.loc[data['Hall number'] == symm["hall_number"]].reset_index(drop=True)
newdata.head()
```
